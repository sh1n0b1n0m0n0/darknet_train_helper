# Make trainset by folders 

```python
import pandas as pd
import numpy as np
from pathlib import Path
import random
from sklearn.model_selection import train_test_split
```

```python
proj_names = ['cropped_349', 
              'cropped_350', 
              'cropped_351', 
              'cropped_352',
              'cropped_353',
              'cropped_354',
              'cropped_355',
              'cropped_356',
              'cropped_357',
              'cropped_358',
              'cropped_359',
              'cropped_360',
              'cropped_361',
              'cropped_362',
              'cropped_363',
              'cropped_364',
              'cropped_365',
              "173",
              "empty_45"
              ]

types = ["*.jp*g", "*.png"]
proj_path = '/home/alexsh/darknet_experiments/data'
data = []

for proj in range(len(proj_names)):
    for t in types:
        try:
            for img in (Path(proj_path) / proj_names[proj]).glob(t):
                # print(str(img).split('/home/alexsh/darknet_experiments/')[1])
                data.append(str(img).split('/home/alexsh/darknet_experiments/')[1])
        except ValueError:
            print("Dataset is Empty!")
```

```python
df = pd.DataFrame(data, columns=["imgs"])
```

```python
df
```

```python
df[df.duplicated(['imgs'], keep=False)]
```

```python
df_train, df_valid = train_test_split(df, train_size=0.80, random_state=42)
```

```python
df_train
```

```python
df_valid
```

```python
train_set_list = df_train["imgs"].tolist()
valid_set_list = df_valid["imgs"].tolist()
```

```python
train_txt_dir = Path(proj_path) / "train_173.txt"
valid_txt_dir = Path(proj_path) / "valid_173.txt"

with open(str(train_txt_dir), "w") as f:
    for img in train_set_list:
        f.write(str(img))
        f.write("\n")

with open(str(valid_txt_dir), "w") as f:
    for img in valid_set_list:
        f.write(str(img))
        f.write("\n")
```

# Дополнительное перемешивание датасетов.

```python
train_txt = Path("/home/alexsh/darknet_experiments/data/default_prepared/t.txt")
valid_txt = Path("/home/alexsh/darknet_experiments/data/default_prepared/v.txt")
```

```python
with open(train_txt, "r") as f:
    train_txt_guts = f.read().splitlines()

with open(valid_txt, "r") as g:
    valid_txt_guts = g.read().splitlines()
```

```python
df_train = pd.DataFrame(train_txt_guts, columns=["imgs"])
df_valid = pd.DataFrame(valid_txt_guts, columns=["imgs"])
```

```python
df_train = df_train.sample(frac=1).reset_index(drop=True)
```

```python
df_valid = df_valid.sample(frac=1).reset_index(drop=True)
```

```python
train_set_list = df_train["imgs"].tolist()
valid_set_list = df_valid["imgs"].tolist()
```

```python
proj_path = '/home/alexsh/darknet_experiments/data/default_prepared'

train_txt_dir = Path(proj_path) / "train.txt"
valid_txt_dir = Path(proj_path) / "valid.txt"

with open(str(train_txt_dir), "w") as f:
    for img in train_set_list:
        f.write(str(img))
        f.write("\n")

with open(str(valid_txt_dir), "w") as f:
    for img in valid_set_list:
        f.write(str(img))
        f.write("\n")
```
